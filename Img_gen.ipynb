```python
from mistralai import Mistral
ob1 = Mistral(api_key)
```

```python
res=ob1.beta.agents.create(
    model="mistral-medium-2505",
    name="Image Generation Model",
    description="Agent used to generate images",
    tools=[{"type": "image_generation"}],
    completion_args={
        "temperature":0.8,
        "top_p":1.0
    }
)
response=ob1.beta.conversations.start(
    agent_id=res.id,
    inputs="Rahul Gandhi winning"
)
```

```python
from mistralai.models import ToolFileChunk
for i,chunk in enumerate(response.outputs[-1].content):
  if isinstance(chunk,ToolFileChunk):
    file_bytes=ob1.files.download(file_id=chunk.file_id).read()
    with open(f"image{i}.png","wb") as f:
      f.write(file_bytes)
```
