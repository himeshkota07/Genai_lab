```python
RCB='''A stampede during RCB's victory parade in Bengaluru on June 4 claimed 11 lives and left several others with injuries, turning a moment of triumph into tragedy.
Facing a backlash for mismanagement and hurried planning, the Congress government in Karnataka suspended top police officials'''

from mistralai import Mistral
client=Mistral(api_key="Yj3vkoFDinDn6W3Dm1nkDiy1BOJbD8Ms")

df3=client.chat.complete(
        model="mistral-tiny",
        messages=[
                   {
                       "role":"user",
                       "content":"can you find the keypoints in the given text: "+RCB
                       }
                   ]
        )
keypoints=df3.choices[0].message.content
print("Keypoints:", keypoints)

df4=client.chat.complete(
        model="mistral-tiny",
        messages=[
                   {
                       "role":"user",
                       "content":"can you summarize the keypoints in the given text: "+keypoints
                       }
                   ]
        )
summary=df4.choices[0].message.content
print("\nSummary:", summary)

df5=client.chat.complete(
        model="mistral-tiny",
        messages=[
                    {
                        "role":"user",
                        "content":"can you do sentiment analysis for each keypoint in the given text: "+summary
                        }
                      ]
        )
sentiment=df5.choices[0].message.content
print("\nSentiment Analysis:", sentiment)
```

```python
def ask_llm(content,text=" "):
  response=client.chat.complete(
        model="mistral-tiny",
        messages=[
                    {
                        "role":"user",
                        "content":content +text
                        }
                      ]
        )
  return response

content="How mant lives were lost in the given event?"
df6=ask_llm(content,summary)
print(df6.choices[0].message.content)
```
