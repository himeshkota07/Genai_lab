```python
def bnm(topic,no_of_question):
  prompt=f"""
  Can you please create {no_of_question} question and answers about the topic {topic} in the format
  <Q>:[Question]
  <A>:[Answer]
  """

  result=ob1.chat.complete(
    model="mistral-tiny",
    messages=[
        {"role": "user", "content": prompt}
    ],
    temperature=0.5,
    top_p=1.0,
    max_tokens=1024
  )
  return result.choices[0].message.content


print(bnm("Pablo Emilio Escobar Gaviria",7))
```
